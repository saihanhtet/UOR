**Question 5: What is the distribution of total vaccinations per 100 across countries?**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vacc_df = pd.read_csv('../../dataset/Vaccination-data.csv',index_col=False)
```

```python
vacc_df.shape
```

```python
# Initialize the figure
plt.figure(figsize=(12, 8))
# Plot a histogram for total vaccinations per 100 people across countries
sns.histplot(vacc_df['TOTAL_VACCINATIONS_PER100'].fillna(0), bins=20, color='green', kde=True)
# Add axis labels and chart title
plt.xlabel('Total Vaccinations per 100', fontsize=12)
plt.ylabel('Number of Countries', fontsize=12)
plt.title('Distribution of Total Vaccinations per 100 Across Countries', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
# Display the plot and save the plot
plt.tight_layout()
plt.savefig('Distribution of Total Vaccinations per 100 Across Countries.png')
plt.show()
plt.close()
```

```python
plt.figure(figsize=(8, 12))
sns.violinplot(y=vacc_df['TOTAL_VACCINATIONS_PER100'].fillna(0), inner='box', color='green')
plt.ylabel('Total Vaccinations per 100', fontsize=12)
plt.title('Distribution of Total Vaccinations per 100 Across Countries (Violin Plot)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('alt_plot5.png')
plt.show()
plt.close()
```
